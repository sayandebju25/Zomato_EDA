# Step 1 - Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# step 2- Create the deta frame

```python
dataframe = pd.read_csv("/content/Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

# Convert the data type of column - rate

```python
def hendlaRate(value):

  value = str(value).split("/")
  value = value[0]
  return float(value)

dataframe['rate']=dataframe['rate'].apply(hendlaRate)
print(dataframe.head())
```

```python
dataframe.info()
```

#1) What type of restaurant do the majority of customers order from?

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of Restaurant")
plt.ylabel("Count")
```

**conclusion - mejority of the resturent falls in dinning category**

#2 How many votes has each type of restaurant received from customers?

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes' : (grouped_data)})
plt.plot(result,c="green",marker="o")
plt.xlabel("Type of Restaurant",c="red",size=20)
plt.ylabel("Votes",c="red",size=20)
```

**conclusion - dinning resturent has recieved maximum votes**

# 3) What are the ratings that the majority of restaurants have received?

```python
dataframe.head()
```

```python
plt.hist(x=dataframe['rate'],bins=5)
plt.title("Rating Distribution")
plt.show()
```

**conclusion - the majority resturant received rating from 3.5 to 4**

#4)Zomato has observed that most couples order most of their food online. What is their average spending on each order?

```python
dataframe.head()
```

```python
couple_data=dataframe["approx_cost(for two people)"]
sns.countplot(x=couple_data)

```

**conclusion - the mejority of coples preferr resturants with an approximate cost of 300 rupees**

#5) Which mode (online or offline) has received the maximum rating?

```python
dataframe.head()
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(x='online_order',y='rate',data=dataframe)
```

**conclusion - offline order receved lower rating in comparison to online order**

# 6) Which type of restaurant received more offline orders, so that Zomato can provide those customers with some good offers?

```python
dataframe.head()
```

```python
pivot_table=dataframe.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap='YlGnBu',fmt='d')
plt.title('Heatmap')
plt.xlabel('Online Order')
plt.ylabel('listed_in(type)')
```

**CONCLUSION: Dining resturants primarily accept offline orders, whereas cafes primarily receve online order.This suggest that clients prefers to order in person at restaurants, but online ordering at cafes.**
